```python
import requests

token = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()
host = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiUrl().get()

response = requests.post(
    f"{host}/api/2.1/unity-catalog/catalogs",
    headers={"Authorization": f"Bearer {token}"},
    json={
        "name": "energy",
        "storage_root": "abfss://datalake@theportfoliostorage.dfs.core.windows.net/",
        "comment": "UK energy carbon intensity DE portfolio project"
    }
)
print(response.status_code)
print(response.json())
```

```python
schemas = ["bronze", "silver", "gold", "serving"]

for schema in schemas:
    response = requests.post(
        f"{host}/api/2.1/unity-catalog/schemas",
        headers={"Authorization": f"Bearer {token}"},
        json={
            "name": schema,
            "catalog_name": "energy",
            "comment": f"Energy pipeline {schema} layer"
        }
    )
    print(f"{schema}: {response.status_code}")
```
